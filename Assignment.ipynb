```python
def binarySearchAlgorithm(array, target):
    return binarySearch(array, 0, len(array) - 1, target)

def binarySearch(array, l, r, target):
    while l <= r:
        mid = (l + r) // 2
        if array[mid] == target:
            return mid
        elif array[mid] < target:
            l = mid + 1
        else:
            r = mid - 1
    return -1

array = [5,10,18,21,32,45,67]
print(binarySearchAlgorithm(array, 45))

```

```python
def bubble_sort(arr):


    n = len(arr)

    # Outer loop: number of passes
    for i in range(n):
        # Inner loop: compare adjacent elements
        for j in range(0, n - i - 1):
            # If current element > next element, swap them
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]

    return arr

# Test the function
arr = [21, 10, 45, 32, 67, 5, 18]
print(f"Original array: {arr}")

sorted_arr = bubble_sort(arr.copy())  # Use copy to preserve original
print(f"Sorted array: {sorted_arr}")
```

```python
# ==================== NODE CLASS ====================
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

# ==================== LINKED LIST CLASS ====================
class LinkedList:
    def __init__(self):
        self.head = None

    def insert_at_beginning(self, data):
        new_node = Node(data)
        new_node.next = self.head
        self.head = new_node
        print(f"Inserted {data} at the beginning")

    def insert_at_end(self, data):
        new_node = Node(data)

        if self.head is None:
            self.head = new_node
            print(f"Inserted {data} at the end (first node)")
            return

        current = self.head
        while current.next is not None:
            current = current.next

        current.next = new_node
        print(f"Inserted {data} at the end")

    def display(self):
        if self.head is None:
            print("List is empty")
            return

        print("Linked List:", end=" ")
        current = self.head
        while current is not None:
            print(current.data, end=" -> ")
            current = current.next
        print("None")

# ==================== TESTING THE LINKED LIST ====================
print("=" * 60)
print("QUESTION FOUR: Linked Lists Implementation")
print("=" * 60)

# Create a new linked list
ll = LinkedList()

print("\n1. Testing insert at beginning:")
ll.insert_at_beginning(10)
ll.display()

ll.insert_at_beginning(20)
ll.display()

ll.insert_at_beginning(30)
ll.display()

print("\n2. Testing insert at end:")
ll.insert_at_end(40)
ll.display()

ll.insert_at_end(50)
ll.display()

print("\n3. Final Linked List:")
ll.display()

print("\n" + "=" * 60)
```

```python
# ==================== STUDENT QUEUE CLASS ====================
class StudentQueue:
    def __init__(self):
        self.queue = []

    def add_student(self, student_name):
        self.queue.append(student_name)
        print(f"✓ {student_name} has joined the queue")

    def serve_student(self):
        if self.is_empty():
            print("✗ Queue is empty! No students to serve.")
            return None

        served_student = self.queue.pop(0)
        print(f"✓ Serving: {served_student}")
        return served_student

    def display_queue(self):
        if self.is_empty():
            print("Queue is empty")
            return

        print("\n--- Current Queue at Finance Office ---")
        print("Position | Student Name")
        print("-" * 35)
        for i, student in enumerate(self.queue, 1):
            print(f"   {i}     | {student}")
        print(f"\nTotal students waiting: {len(self.queue)}")
        print("-" * 35)

    def is_empty(self):
        return len(self.queue) == 0

    def size(self):
        return len(self.queue)


# ==================== TESTING THE SYSTEM ====================
print("=" * 60)
print("QUESTION FIVE: Finance Office Queue Management System")
print("=" * 60)

# Create a new queue
finance_queue = StudentQueue()

print("\n--- Scenario: Students Arriving at Finance Office ---\n")

# Students joining the queue
print("1. Students joining the queue:")
finance_queue.add_student("Alice")
finance_queue.add_student("Bob")
finance_queue.add_student("Charlie")
finance_queue.add_student("Diana")
finance_queue.add_student("Eve")

# Display current queue
finance_queue.display_queue()

print("\n2. Serving students:")
# Serve first student
finance_queue.serve_student()
finance_queue.display_queue()

# Serve second student
finance_queue.serve_student()
finance_queue.display_queue()

print("\n3. More students joining:")
# New students arrive
finance_queue.add_student("Frank")
finance_queue.add_student("Grace")
finance_queue.display_queue()

print("\n4. Serving remaining students:")
# Serve all remaining students
while not finance_queue.is_empty():
    finance_queue.serve_student()

# Try to serve when queue is empty
print("\n5. Attempting to serve when queue is empty:")
finance_queue.serve_student()

print("\n" + "=" * 60)
```
